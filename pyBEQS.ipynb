## EQS: Run a EQS screen and capture output tickers

```python
import blpapi
import pandas as pd
```

### Function definition

```python
def getData(screenname):
    session=blpapi.Session()
    session.start()
    session.openService("//blp/refdata")
    
    refDataService=session.getService("//blp/refdata")
    request=refDataService.createRequest("BeqsRequest")
    
    # Custom, Personal EQS Screen Name
    request.set("screenName", screenname) 
    request.set("screenType", "PRIVATE")
    session.sendRequest(request)

    response={}
    while(True):
        event = session.nextEvent()
        if event.eventType() ==blpapi.Event.RESPONSE or event.eventType() ==blpapi.Event.PARTIAL_RESPONSE:
            response={}
            for msg in event:
                secData = msg.getElement("data").getElement("securityData").values()
                for Sec in secData:
                    security = Sec.getElement("security").getValue()
                    response[security]={}
                    FldsData=Sec.getElement("fieldData")            
                    numFldsEls=FldsData.numElements()
                    response[security]["secData"]=[]
                    for i in range(1,numFldsEls-1):
                        response[security]["secData"].append(str(FldsData.getElement(i)))

        if event.eventType() == blpapi.Event.RESPONSE:
            break          
    return response
                           
```

#### Execute Function store results in 'results'

```python
# wrapper of getData function to return a datafram
def getEQSData(screenname):
    results=getData(screenname)  #screename is the argument to getData()
    tickerlist = pd.DataFrame(results).T
    tickers =tickerlist.reset_index()[[0]]
    tickers.rename(columns= {'index':'ticker'},inplace=True)
    return tickers
```

#### Example call of `getEQSData()`

```python
eqstickers = getEQSData("CCNY")
```
